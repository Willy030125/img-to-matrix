```python
import sys
import numpy as np
import pandas as pd
import tensorflow as tf
from tensorflow.keras.utils import load_img
from tensorflow.image import rgb_to_grayscale
import matplotlib.pyplot as plt

np.set_printoptions(threshold=sys.maxsize)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
pd.set_option('display.width', None)
pd.set_option('display.max_colwidth', None)
```

```python
img = load_img("8.jpg", target_size=(50,50))
grayscale = rgb_to_grayscale(img)
img = np.asarray(grayscale)
plt.imshow(img, cmap='gray')
```

```python
img.shape
```

```python
m = []
for pixel in img:
    n = np.array([])
    for inside_pixel in pixel:
        value = inside_pixel
        # invert the color, so it's easier to read the matrix
        invert = np.array([255]) - value
        n = np.append(n, invert)
    m.append(n)
```

```python
x = np.array(m, dtype=int)
printout = pd.DataFrame(x)
printout
```
